```python
import pandas as pd
import numpy as np

file_path = 'Data/cumulative_2025.10.04_11.31.09.csv'  

df = pd.read_csv(file_path, comment='#')

print("Header:", df.columns.tolist())
print("First 5 rows of data:", df.head())

# check for confirmed vs false positive in koi_pdisposition
print(df['koi_disposition'].value_counts())
```

```python
print( df.columns.tolist())
```

```python
# keep only binary classification: CANDIDATE vs non-CANDIDATE
df_binary = df[df['koi_disposition'] != 'CANDIDATE']

print("Original DataFrame size:", df.shape)
print("Filtered DataFrame size:", df_binary.shape)

df_binary.head()
```

```python
# keep only kepid, koi_disposition, koi_period, koi_time0bk, koi_impact, koi_duration, koi_depth, koi_prad, koi_teq, koi_insol, koi_model_snr, koi_steff, koi_slogg, koi_srad, ra, dec, koi_kepmag
df_columns_of_interest = ['kepid', 'koi_disposition', 'koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 'koi_steff', 'koi_slogg', 'koi_srad', 'ra', 'dec', 'koi_kepmag']

df_filtered = df_binary[df_columns_of_interest]

df_filtered.head()
```

```python
# check for null values and empty columns
print("Null values in each column:\n", df_filtered.isnull().sum())

# show confirmed vs false positive
print(df_filtered['koi_disposition'].value_counts())

# remove rows with null values
df_filtered_dropped_null = df_filtered.dropna()

# show the size of the dataframe after removing null values
print("\n DataFrame size after removing null values:", df_filtered_dropped_null.shape)
print(df_filtered_dropped_null['koi_disposition'].value_counts())
```

```python
# add column for binary classification: 1 for CONFIRMED, 0 for FALSE POSITIVE using loc
df_filtered_dropped_null.loc[df_filtered_dropped_null['koi_disposition'] == 'CONFIRMED', 'label'] = 1
df_filtered_dropped_null.loc[df_filtered_dropped_null['koi_disposition'] == 'FALSE POSITIVE', 'label'] = 0

# drop koi_disposition column keep only label column
df_binary_labeled = df_filtered_dropped_null.drop(columns=['koi_disposition'], axis=1)

df_binary_labeled.head()
```

```python
# save as new csv file
df_binary_labeled.to_csv('Data/preprocessed_exoplanet_data.csv', index=False)
```
